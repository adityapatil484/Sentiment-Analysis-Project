```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime 
```

# Importing data from Json files

```python
#importing all the Nike data from Json file

file = "nike.json"
data = None
with open(file, 'r') as f:
    data = json.load(f)
    
channel_id, stats = data.popitem()


print(channel_id)

channel_stats = stats["channel_statistics"]
video_stats = stats["video_data"]
```

```python
#importing all the Puma data from Json file

file = "PUMA.json"
data = None
with open(file, 'r') as f:
    data = json.load(f)
    
channel_id2, stats2 = data.popitem()

print(channel_id2)

channel_stats2 = stats2["channel_statistics"]
video_stats2 = stats2["video_data"]
```

```python
#importing all the Adidas data from Json file

file = "adidas.json"
data = None
with open(file, 'r') as f:
    data = json.load(f)
    
channel_id3, stats3 = data.popitem()

print(channel_id3)

channel_stats3 = stats3["channel_statistics"]
video_stats3 = stats3["video_data"]
```

# Finding Channel Stats

```python
# Nike channel statsistics 

print('views', channel_stats['viewCount'])
print('subscriber', channel_stats['subscriberCount'])
print('videos', channel_stats['videoCount'])

nike_total_views = int(channel_stats['viewCount'])
nike_subscriber = int(channel_stats['subscriberCount'])
nike_total_videos = int(channel_stats['videoCount'])
```

```python
# Puma channel statsistics 

print('total_views', channel_stats2['viewCount'])
print('subscriber', channel_stats2['subscriberCount'])
print('videos', channel_stats2['videoCount'])

puma_total_views = int(channel_stats2['viewCount'])
puma_subscriber = int(channel_stats2['subscriberCount'])
puma_total_videos = int(channel_stats2['videoCount'])
```

```python
# Adidas channel statsistics 

print('views', channel_stats3['viewCount'])
print('subscriber', channel_stats3['subscriberCount'])
print('videos', channel_stats3['videoCount'])


adidas_total_views = int(channel_stats3['viewCount'])
adidas_subscriber = int(channel_stats3['subscriberCount'])
adidas_total_videos = int(channel_stats3['videoCount'])
```

# Creating data frames

```python
# Nike video stats into dataframe

nike_vids = list(video_stats.items())
stats = []
for vid in nike_vids:
    video_id = vid[0]
    title = vid[1]['title']
    tags = vid[1]['tags']
    views = int(vid[1]['viewCount'])
    likes = int(vid[1]['likeCount'])
    dislikes = int(vid[1]['dislikeCount'])
    comments = int(vid[1]["commentCount"])
    date = (vid[1]["publishedAt"])
    stats.append([title,tags, views, likes, dislikes, comments,date])
    
nike_df = pd.DataFrame(stats, columns = ["title", "tags", "views","likes","dislikes","comments","date"])


nike_df['date']= pd.to_datetime(nike_df['date'])

nike_df['Date'] = [datetime.datetime.date(d) for d in nike_df['date']]
nike_df['Time'] = [datetime.datetime.time(d) for d in nike_df['date']]

nike_df = nike_df.drop('date', axis=1)



nike_df.head()
```

```python
# Puma video stats into dataframe

puma_vids = list(video_stats2.items())
stats2 = []
for vid in puma_vids:
    video_id = vid[0]
    title = vid[1]['title']
    views = int(vid[1]['viewCount'])
    likes = int(vid[1]['likeCount'])
    dislikes = int(vid[1]['dislikeCount'])
    comments = int(vid[1]["commentCount"])
  
    date = (vid[1]["publishedAt"])
    stats2.append([title, views, likes, dislikes, comments,date])
    
puma_df = pd.DataFrame(stats2, columns = ["title", "views","likes","dislikes","comments","date"])

puma_df['date']= pd.to_datetime(puma_df['date'])


puma_df['Date'] = [datetime.datetime.date(d) for d in puma_df['date']]
puma_df['Time'] = [datetime.datetime.time(d) for d in puma_df['date']]

puma_df = puma_df.drop('date', axis=1)

puma_df.head()

```

```python
# Adidas video stats into dataframe

adidas_vids = list(video_stats3.items())
stats3 = []
for vid in adidas_vids:
    video_id = vid[0]
    title = vid[1]['title']
    views = int(vid[1]['viewCount'])
    likes = int(vid[1]['likeCount'])
    dislikes = int(vid[1]['dislikeCount'])
    comments = int(vid[1]['commentCount'])
    date = vid[1]["publishedAt"]
    stats3.append([title, views, likes, dislikes, comments,date])
    
adidas_df = pd.DataFrame(stats3, columns = ["title","views","likes","dislikes","comments","date"])

adidas_df['date']= pd.to_datetime(adidas_df['date'])


adidas_df['Date'] = [datetime.datetime.date(d) for d in adidas_df['date']]
adidas_df['Time'] = [datetime.datetime.time(d) for d in adidas_df['date']]

adidas_df = adidas_df.drop('date', axis=1)

adidas_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(nike_df['dislikes'].values.reshape(-1, 1), nike_df['comments'].values.reshape(-1, 1))
# import the regressor
from sklearn.ensemble import RandomForestRegressor
  
 # create regressor object
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
  
# fit the regressor with x and y data
regressor.fit(X_train, y_train)
Y_pred = regressor.predict(y_test)
from matplotlib.pyplot import figure
figure(figsize=(8, 6), dpi=80)
X_grid = np.arange(min(X_train), max(X_train), 0.01) 
  
# reshape for reshaping the data into a len(X_grid)*1 array, 
# i.e. to make a column out of the X_grid value                  
X_grid = X_grid.reshape((len(X_grid), 1))
  
# Scatter plot for original data
plt.scatter(X_train, y_train, color = 'blue')  
  
# plot predicted data
plt.plot(X_grid, regressor.predict(X_grid), 
         color = 'green') 
plt.title('Random Forest Regression for Dislikes vs Comments for Nike')
plt.xlabel('Dislikes')
plt.ylabel('Comments')
plt.show()
```
